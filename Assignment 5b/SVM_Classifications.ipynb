```python
# Import libraries for the SVM linear classification model
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
import seaborn as sns
```

```python
# Create a dataset
X, y = make_moons(n_samples=100, noise=0.2, random_state=42)
```

```python
# Fit the SVM Model
from sklearn.svm import SVC

clf = SVC(kernel='linear', C=1.0)
clf.fit(X, y)
```

```python
# Create a mesh to plot the decision boundary
xx, yy = np.meshgrid(np.linspace(X[:, 0].min()-1, X[:, 0].max()+1, 500),
                     np.linspace(X[:, 1].min()-1, X[:, 1].max()+1, 500))
```

```python
# Plot the decision boundary
Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
```

```python
Z = Z.reshape(xx.shape)
```

```python
plt.figure(figsize=(10, 6))
sns.set(style='whitegrid')

# Plot the decision boundary
plt.contourf(xx, yy, Z, levels=[Z.min(), 0, Z.max()], alpha=0.2, colors=['#FFAAAA', '#AAAAFF'])
plt.contour(xx, yy, Z, levels=[0], linewidths=2, colors=['k'])

# Plot the data points
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y, palette='coolwarm', edgecolor='k', s=100)

# Plot support vectors
plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=200,
            facecolors='none', edgecolors='k', linewidths=1.5, label='Support Vectors')

plt.title('SVM with Linear Kernel')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
# Fit the polynomial SVM Model
clf_poly = SVC(kernel='poly', degree=3, C=1000.0)
clf_poly.fit(X, y)
```

```python
# Create a mesh to plot the decision boundary
xx, yy = np.meshgrid(np.linspace(X[:, 0].min()-1, X[:, 0].max()+1, 500),
                     np.linspace(X[:, 1].min()-1, X[:, 1].max()+1, 500))
```

```python
# Plot the decision boundary
Z = clf_poly.decision_function(np.c_[xx.ravel(), yy.ravel()])
```

```python
Z = Z.reshape(xx.shape)
```

```python
plt.figure(figsize=(10, 6))
sns.set(style='whitegrid')

# Plot the decision boundary
plt.contourf(xx, yy, Z, levels=[Z.min(), 0, Z.max()], alpha=0.2, colors=['#FFAAAA', '#AAAAFF'])
plt.contour(xx, yy, Z, levels=[0], linewidths=2, colors=['k'])

# Plot the data points
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y, palette='coolwarm', edgecolor='k', s=100)

# Plot support vectors
plt.scatter(clf_poly.support_vectors_[:, 0], clf_poly.support_vectors_[:, 1], s=200,
            facecolors='none', edgecolors='k', linewidths=1.5, label='Support Vectors')

plt.title('SVM with Polynomial Kernel')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```
